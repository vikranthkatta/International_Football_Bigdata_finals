```python
# "clean_results"
df = spark.table("international_football_matches.default.clean_results")

df.show(5)
```

```python
# Basic stats
df.describe(["home_score", "away_score", "total_goals", "goal_difference"]).show()
```

```python
# top 10 teams by win
top_teams = (
    df.filter(df.winner != "Draw")
      .groupBy("winner")
      .count()
      .orderBy("count", ascending=False)
)

top_teams.show(10)
```

```python
from pyspark.sql.functions import floor

df_with_decade = df.withColumn("decade", (floor(df.year / 10) * 10))

avg_goals_decade = (
    df_with_decade
    .groupBy("decade")
    .avg("total_goals")
    .orderBy("decade")
)

avg_goals_decade.show()
```
