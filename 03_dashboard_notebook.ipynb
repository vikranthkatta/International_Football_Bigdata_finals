```python
try:
    df = spark.table("international_football_matches.default.clean_results")
except:
    df = spark.table("clean_results")

df.show(5)
```

```python
df_top_wins = (
    df.filter(df.winner != "Draw")
      .groupBy("winner")
      .count()
      .orderBy("count", ascending=False)
      .limit(10)
)

display(df_top_wins)
```

```python
from pyspark.sql.functions import avg

df_avg_goals = (
    df.groupBy("year")
      .agg(avg("total_goals").alias("avg_goals_per_match"))
      .orderBy("year")
)

display(df_avg_goals)
```

```python
df_result_dist = (
    df.groupBy("result")
      .count()
      .orderBy("count", ascending=False)
)

display(df_result_dist)
```

```python
df_top_wins.write.mode("overwrite").saveAsTable("international_football_matches.default.dash_top_wins")
df_avg_goals.write.mode("overwrite").saveAsTable("international_football_matches.default.dash_avg_goals")
df_result_dist.write.mode("overwrite").saveAsTable("international_football_matches.default.dash_result_dist")
```
